```python
data.columns = (
    data.columns
    .str.strip()          # hapus spasi awal & akhir
    .str.lower()          # ubah ke huruf kecil
    .str.replace(" ", "_")# ganti spasi jadi underscore
)
```

```python
data_2.columns = (
    data_2.columns
    .str.strip()          # hapus spasi awal & akhir
    .str.lower()          # ubah ke huruf kecil
    .str.replace(" ", "_")# ganti spasi jadi underscore
)
```

```python
data_3 = pd.read_csv('MentalHealthSurvey.csv')
data_3.head()
```

```python
data_3.columns = (
    data_3.columns
    .str.strip()          # hapus spasi awal & akhir
    .str.lower()          # ubah ke huruf kecil
    .str.replace(" ", "_")# ganti spasi jadi underscore
)
```

```python
print(data.columns), (data_2.columns), (data_3.columns)
```

```python
print("Missing Values Dataset 1:")
print(data.isna().sum())

print("\nMissing Values Dataset 2:")
print(data_2.isna().sum())

print("\nMissing Values Dataset 3:")
print(data_3.isna().sum())
```

##### Normalisasi gender

```python
for i, df in enumerate([data, data_2, data_3], start=1):
    if "gender" in df.columns:
        print(f"\nDataset {i} - unique gender:")
        print(df["gender"].unique())
```

```python
def normalize_gender(series):
    return (
        series
        .astype(str)
        .str.strip()
        .str.lower()
        .replace({
            # Dataset 1 (binary)
            "1": "male",
            "0": "female",

            # Male variants
            "male": "male",

            # Female variants
            "female": "female",

            # Other / undisclosed
            "non-binary": "unknown",
            "prefer not to say": "unknown",
            "nan": "unknown",
            "none": "unknown"
        })
    )
```

```python
for df in [data, data_2, data_3]:
    if "gender" in df.columns:
        df["gender"] = normalize_gender(df["gender"])
```

```python
for i, df in enumerate([data, data_2, data_3], start=1):
    if "gender" in df.columns:
        print(f"\nDataset {i} - unique gender:")
        print(df["gender"].unique())
```

##### Normalisasi depresi menjadi biner

```python
PHQ9_THRESHOLD = 10

data["depression_binary"] = (data["depression_score"] >= PHQ9_THRESHOLD).astype(int)
data_2["depression_binary"] = (data_2["depression_score"] >= PHQ9_THRESHOLD).astype(int)
```

```python
data_3.rename(columns={"depression": "depression_binary"}, inplace=True)
```

```python
data_3["depression_binary"] = data_3["depression_binary"].apply(
    lambda x: 1 if x >= 3 else 0
)
```

```python
for i, df in enumerate([data, data_2, data_3], start=1):
    if "depression_binary" in df.columns:
        print(f"\nDataset {i} - nilai unik depression_binary:")
        print(df["depression_binary"].unique())
    else:
        print(f"\nDataset {i} - kolom depression_binary belum ada")
```

##### Integrasi data1 dan 2 

```python
data_1_2 = pd.concat([data, data_2], ignore_index=True)
```
